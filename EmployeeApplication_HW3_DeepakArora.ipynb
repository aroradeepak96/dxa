```python
class Employee:

    def __init__(self, ename, eaddress,make,model,manuyear,mileage):
        #self.__employee_type = etype
        self.__employee_name = ename
        self.__employee_address = eaddress
        self.__employee_vehicle = Vehicle(make,model,manuyear,mileage)

    def get_name(self):
        return self.__employee_name
    def get_address(self):
        return self.__employee_address
    def get_vehicle(self):
        return self.__employee_vehicle
    def set_name(self,ename):
        self.__employee_name= ename
    def set_address(self,eaddress):
        self.__employee_address= eaddress
    def set_vehicle(self,evehicle):
        self.__employee_vehicle= Vehicle(make,model,manuyear,mileage)

    
```

```python
class Full_Time_Employee(Employee):
    def __init__(self, ename,eaddress,make,model,manuyear,mileage,esalary):
        super().__init__(ename,eaddress,make,model,manuyear,mileage) #initializing the instance variables of the base class
        self.__esalary = esalary
    #def __init__(self, Sal):
        #self.__esalary = Sal

    def get_esalary(self):
        return self.__esalary

    def set_esalary(self,esalary):
        self.__esalary = esalary

    def compute_compensation(self):
        nc = self.get_esalary()
        if nc <= 45000:
            tax= (nc * 18)/100
        elif nc> 45000 and nc<= 82000:
            tax = (45000 * 18)/100 + ((nc -45000)* 28)/100
        elif nc>82000:
            tax = ((45000 * 18)/100 + ((82000 -45000)* 28)/100 + ((nc-82000)* 33)/100)
        else:
            print(" invalid Salary!!")
        return (nc-tax)
        
```

```python
class Hourly_Employee(Employee):
    def __init__(self, ename,eaddress,make,model,manuyear,mileage,ehour,erate):
        super().__init__(ename,eaddress,make,model,manuyear,mileage) 
        self.__ehour = ehour
        self.__erate = erate
    def get_ehour(self):
        return self.__ehour

    def set_ehour(self, ehour):
        self.__ehour = ehour

    def get_erate(self):
        return self.__erate
                
    def set_erate(self, erate):
        self.__erate = erate
        
    def compute_compensation(self):
        hrs = self.get_ehour()
        rate= self.get_erate()
        if hrs > 40:
            return (40*rate)+(rate*(hrs-40)*1.8)
        else:
            return hrs*rate
```

```python
class Consultant(Employee):
    def __init__(self, ename,eaddress,make,model,manuyear,mileage,ehour1,eproject):
        super().__init__(ename,eaddress,make,model,manuyear,mileage) #initializing the instance variables of the base class
        self.__ehourworked = ehour1
        self.__eproject = eproject
    def get_ehourworked(self):
        return self.__ehourworked
            
    def set_ehourworked(self, ehour1):
        self.__ehourworked = ehour1

    def get_eproject(self):
        return self.__eproject

    def set_eproject(self, ProjectType ):
        self.__eproject = ProjectType
    
    def compute_compensation(self):
        pt = self.get_eproject()
        if pt == 1:
            return 55 * self.get_ehourworked()
        if pt == 2:
            return 70 * self.get_ehourworked()
        if pt == 3:
            return 85 * self.get_ehourworked()
```

```python
class Vehicle(Employee):
    def __init__(self, make, model, manuyear,mileage):
        self.__make = make
        self.__model = model
        self.__manuyear = manuyear
        self.__mileage = mileage
    def get_make(self):
        return self.__make

    def set_make(self, make):
        self.__make = make

    def get_model(self):
        return self.__model

    def set_model(self, model ):
        self.__model = model
    def get_manuyear(self):
        return self.__manuyear
            
    def set_manuyear(self, manuyear ):
        self.__manuyear = manuyear
    
    def get_mileage(self):
        return self.__mileage

    def set_mileage(self, mileage ):
        self.__mileage = mileage
```
